Problem 1: Mobile Recharge Validation System


```python


def validate_recharge(amount):
    valid_plans = [199, 299, 399, 599]

    if amount < 50:
        print("Recharge must be at least ₹50")
        return False
    elif amount not in valid_plans:
        print("Invalid plan. Valid plans are:", valid_plans)
        return False
    else:
        print("Recharge Successful!")
        return True

# Retry option
while True:
    amount = int(input("Enter recharge amount: "))
    if validate_recharge(amount):
        break
```

Problem 2: Inventory Reorder Alert System


```python


def check_inventory(products):
    for product, stock in products.items():
        if stock < 15:
            print(f"{product}: Reorder Alert (Stock = {stock})")
        else:
            print(f"{product}: Stock OK (Stock = {stock})")

inventory = {
    "Apples": 10,
    "Bananas": 25,
    "Oranges": 12,
    "Mangoes": 30
}

check_inventory(inventory)
```

Problem 3: Student Result Processing System

```python

def process_result(marks):
    total = 0
    for m in marks:
        total += m

    average = total / len(marks)
    print("Average Marks:", average)

    if average >= 50:
        return "Pass"
    else:
        return "Fail"

marks = [45, 60, 55, 70]
result = process_result(marks)
print("Result:", result)
```

Problem 4: Cab Fare Estimator with Retry Option


```python

def calculate_fare(km, peak):
    base_fare = 50
    fare = base_fare + (12 * km)

    if peak.lower() == "yes":
        fare *= 1.25   # 25% extra

    return fare

# Retry option
while True:
    km = float(input("Enter distance (km): "))
    peak = input("Is it peak hour? (yes/no): ")

    fare = calculate_fare(km, peak)
    print("Estimated Fare: ₹", round(fare, 2))

    retry = input("Do you want to calculate again? (yes/no): ")
    if retry.lower() != "yes":
        break
```

Problem 5: Employee Attendance Eligibility Checker


```python

def check_eligibility(attendance):
    present = 0

    for day in attendance:
        if day == "P":
            present += 1

    percentage = (present / len(attendance)) * 100
    print("Attendance Percentage:", percentage)

    if percentage >= 75:
        return "Eligible"
    else:
        return "Not Eligible"

attendance = ["P", "P", "A", "P", "P", "A", "P"]
status = check_eligibility(attendance)
print("Eligibility Status:", status)
```

Problem 6: Password Strength Checker


```python

def check_password_strength(password):
    special_chars = "@#$"
    has_digit = False
    has_special = False

    if len(password) < 8:
        return "Weak Password (Minimum length 8 required)"

    for char in password:
        if char.isdigit():
            has_digit = True
        if char in special_chars:
            has_special = True

    if has_digit and has_special:
        return "Strong Password"
    else:
        return "Weak Password (Must contain digit and special character)"

password = input("Enter password: ")
print(check_password_strength(password))
```
